```python
import pandas as pd
import os 
```

```python
resource_path = './Resources'
only_files = [os.path.join(resource_path, f) for f in os.listdir(resource_path) if os.path.isfile(os.path.join(resource_path, f)) 
              and f.find('.csv') != -1]
dfs = []
for f in only_files:
    df = pd.read_csv(f)
    dfs.append(df)

citibike_2019= pd.concat(dfs)
citibike_2019.to_csv("citibike_2019_data.csv", sep=",")
citibike_2019.head()
```
